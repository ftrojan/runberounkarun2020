```python
import datetime
import io
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv(io.StringIO(vysl_data), sep="\t")
```

```python
df1
```

```python
mista = df1.iloc[0, [2,3,5,7,9,11,13]].values
mista
```

```python
tymy = df1.iloc[1:,1]
tymy.index = df1.iloc[1:,0].astype(int)
tymy
```

```python
start = df1.iloc[1:,2]
start.index = tymy
start
```

```python
usek_cas_cil = df1.iloc[1:, [3,5,7,9,11,13]]
usek_cas_cil.index = tymy
usek_cas_cil.columns = range(1,7)
usek_cas_cil
```

```python
def clean_cas(s: str) -> datetime.datetime:
    hhs, mms = s.split(':')
    hh = int(hhs)
    mm = int(mms)
    t = datetime.datetime(2020,10,10,hh,mm)
    return t
```

```python
t_start = start.map(clean_cas)
t_start
```

```python
t_usek_cil = usek_cas_cil.apply(lambda x: x.map(clean_cas), axis=1)
t_usek_cil[0] = t_start
t_usek_cil = t_usek_cil.sort_index(axis=1)
t_usek_cil
```

```python
useky = pd.DataFrame([
    (1, 21.0, 459),
    (2, 19.6, 538),
    (3, 22.3, 588),
    (4, 19.5, 541),
    (5, 23.9, 379),
    (6, 21.8, 493),
], index=range(1,7), columns=['usek', 'delka_km', 'prevyseni_m'])
useky['misto_start'] = mista[0:6]
useky['misto_cil'] = mista[1:7]
useky
```

```python
t_usek_cil.iloc[:, 1:]
```

```python
df = pd.concat([
    pd.melt(t_usek_cil.iloc[:, 0:-1].reset_index(), id_vars='tym', var_name='usek_pred', value_name='t_start').drop(['usek_pred'], axis=1),
    pd.melt(t_usek_cil.iloc[:, 1:].reset_index(), id_vars='tym', var_name='usek', value_name='t_cil').drop(['tym'], axis=1),
], axis=1)
df['t'] = df.t_cil - df.t_start
ds = pd.merge(df, useky, on='usek')
ds['tempo'] = ds.t/ds.delka_km
ds
```
